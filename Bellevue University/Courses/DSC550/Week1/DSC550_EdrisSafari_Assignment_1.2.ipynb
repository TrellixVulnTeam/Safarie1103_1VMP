```python
import pandas as pd
import numpy as np
MAX_ROWS,MAX_COLS =  10,50
```

1. Create an array of size [10,50] and use the random function to fill it with integers between 1 and 500 (inclusive). Load it to a DataFrame.

```python
# Create array fill with randint(1,501)-1 to 500
numRows,numColumns = MAX_ROWS,MAX_COLS
matrix = np.array([[np.random.randint(1,501) for i in range(numColumns)] for j in range(numRows)])
matrix
```

```python
#Examine shpae
matrix.shape
```

```python
# Load matrix to dataframe
dataframe = pd.DataFrame(matrix)
dataframe
```

2. Calculate the sum of each row; the sum of each column; and the sum of all entries.

```python
# Sum of Rows
dataframe.sum(axis=1)
```

```python
# Sum of columns
dataframe.sum(axis=0)
```

```python
# Sum of all entries - Method1
dataframe.sum().sum()
```

```python
# Sum of all entries - Method2
dataframe.sum(axis=1).sum()
```

```python
# Sum of all entries - Method3
dataframe.sum(axis=0).sum()
```

3. Determine the minimum; maximum; average; and standard deviation of all entries.

```python
# Maximum of all entries
dataframe.max().max()
```

```python
# Minimum of all entries
dataframe.min().min()
```

```python
# Average of all entries
dataframe.mean().mean()
```

```python
# standard deviation of all entries
dataframe.std().std()
```

4. Sort the DataFrame: by rows, by columns, on row 2, on column 2.

```python
# Sort by rows on row 2
dataframe.sort_values(by=1,axis=1)
```

```python
# Sort by columns on col 2
dataframe.sort_values(by=1,axis=0)
```
