## Including Library and Paths

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import matplotlib.pyplot as plt # For plots and graphs

import os
for dirname, _, filenames in os.walk('.'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

## Reading the .CSV file

```python
df = pd.read_csv("/kaggle/input/fivethirtyeight-airline-safety-dataset/airline-safety.csv")
```

## Visualizing the dataset

```python
df.head()
```

## Total number of rows in dataframe

```python
print(df.shape[0])
for col in df.columns:
    print(col)
```

## Showing a graph about incidents from year 85 to 99;
## having x axis on airlines count and y on fatalities

```python
plt.figure(figsize=(20,5))
df['incidents_85_99'].value_counts().plot(kind='bar',x='airline',y='incidents_85_99')
plt.xlabel("AirLines count")
plt.ylabel("Incidents happened between year 85 to 99")
plt.show()
```

## Showing a graph about incidents from year 2000 to 2014;
## having x axis on airlines count and y on fatalities

```python
plt.figure(figsize=(20,5))
df['incidents_85_99'].value_counts().plot(kind='bar',x='airline',y='incidents_00_14')
plt.xlabel("AirLines count")
plt.ylabel("Incidents happened between year 2000 to 2014")
plt.show()
```

## Which Airline suffered most incidents in the year from 85 to 99

```python
fig = plt.figure(figsize=(60,5))
ax = fig.add_axes([0,0,1,1])
ax.bar(df['airline'],df['incidents_85_99'])
plt.rcParams.update({'font.size': 13})
plt.xlabel("AirLine")
plt.ylabel("Incidents happened between year 85 to 99")
# uncomment if we want to download
# plt.savefig('plot.png', dpi=300, bbox_inches='tight')
plt.show()
```

## On pie graph

```python
airlines = df["airline"]
number_of_incidents_happened_1985_1999 = df["incidents_85_99"]
fig = plt.figure(figsize =(30, 15))
plt.rcParams.update({'font.size': 10})
plt.pie(number_of_incidents_happened_1985_1999)
plt.legend(title = "AirLines", labels = airlines)
plt.title("Number of incidents happened from 1985 to 1999")
# Uncomment to download
# plt.savefig('plot.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Which Airline suffered most incidents in the year from 2000 to 2014

```python
fig = plt.figure(figsize=(60,5))
ax = fig.add_axes([0,0,1,1])
ax.bar(df['airline'],df['incidents_00_14'])
plt.rcParams.update({'font.size': 10})
plt.xlabel("AirLine")
plt.ylabel("Incidents happened between year 2000 to 2014")
plt.show()
```

## On pie graph

```python
airlines = df["airline"]
number_of_incidents_happened_2000_2014 = df["incidents_00_14"]
fig = plt.figure(figsize =(20, 15))
plt.title("Number of incidents happened from 2000 to 2014")
plt.rcParams.update({'font.size': 10})
plt.pie(number_of_incidents_happened_2000_2014)
plt.legend(title = "AirLines", labels = airlines)
plt.show()
```

## Let's figure out rate of change in incidents from AirLine Aeroflot* from year 85 to 2014

```python
#index of aeroFlot = 1
print(df.iloc[1])
x=["85_99","00_14"]
incidents_happened_y = []
incidents_happened_y.append(df.iloc[1]["incidents_85_99"])
incidents_happened_y.append(df.iloc[1]["incidents_00_14"])
print(x, incidents_happened_y)
```

## Aeroflot AirCrash has decreased over time from the year 1985 to 2014

```python
#Plotting Graph to view changes
plt.plot(x, incidents_happened_y)
plt.xlabel("AirLine AeroFlot* from year 1985 to 2014")
plt.ylabel("Incidents happened between year 2000 to 2014")
plt.show()
```

## Now, let's review the accidents and fatal accidents and fatalities from the year 1985 to 2014

```python
#index of aeroFlot = 1
print(df.iloc[1])
x=["accidents_85_99","fatalities_85_99", "accidents_00_14", "fatalities_00_14"]
incidents_happened_y = []
incidents_happened_y.append(df.iloc[1]["fatal_accidents_85_99"])
incidents_happened_y.append(df.iloc[1]["fatalities_85_99"])
incidents_happened_y.append(df.iloc[1]["fatal_accidents_00_14"])
incidents_happened_y.append(df.iloc[1]["fatalities_00_14"])
print(x, incidents_happened_y)
```

```python
#Plotting Graph to view changes
plt.plot(x, incidents_happened_y)
plt.xlabel("AirLine AeroFlot*")
plt.ylabel("Accidents and Fatalities happened between year 2000 to 2014")
plt.show()
```
