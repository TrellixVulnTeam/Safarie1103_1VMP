```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import xlrd

```

```python
plt.style.use('ggplot')

```

Load the data

```python
worldPopulation = pd.read_excel('Data\world-population.xlsm')
worldPopulation.tail()
```

Python Line  Chart

```python
worldPopulation.set_index('Year').plot(kind='line')
plt.xticks(rotation=90)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population ')
plt.legend(loc='center right', bbox_to_anchor=(1.3, 0.5))
```

```python
worldPopulationByDecade = worldPopulation.loc[worldPopulation['Year'] % 10 ==0 ]
worldPopulationByDecade.head()
```

```python
#worldPopulation.set_index('Year').plot(kind='step')
plt.step(worldPopulationByDecade["Year"], worldPopulationByDecade["Population"], 'r*', where='post')

plt.xticks(rotation=90)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population ')
plt.legend(loc='center right', bbox_to_anchor=(1.3, 0.5))
plt.show()                                                  
```

```python
#worldPopulation.set_index('Year').plot(kind='step')
plt.step(worldPopulation["Year"], worldPopulation["Population"], 'r*', where='post')

plt.xticks(rotation=90)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population ')
plt.legend(loc='right', bbox_to_anchor=(1.3, 0.5))
plt.show()                                                  
```

```python
x = worldPopulation["Year"]
y = worldPopulation["Population"]

plt.step(x, y + 2, label='pre (default)')
plt.plot(x, y + 2, 'o--', color='grey', alpha=0.3)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = worldPopulationByDecade["Year"]
y = worldPopulationByDecade["Population"]
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population ')
plt.step(x, y + 2, label='pre (default)')
plt.plot(x, y + 2, 'o--', color='grey', alpha=0.3)
```

```python

plt.step(x, y + 1, where='mid', label='mid')
plt.plot(x, y + 1, 'o--', color='grey', alpha=0.3)

```

```python
plt.step(x, y, where='post', label='post')
plt.plot(x, y, 'o--', color='grey', alpha=0.3)

plt.grid(axis='x', color='0.95')
plt.legend(title='Parameter where:')
plt.title('plt.step(where=...)')
plt.show()
```
