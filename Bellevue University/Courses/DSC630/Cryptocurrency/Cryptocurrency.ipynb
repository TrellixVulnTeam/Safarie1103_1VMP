```python


import cbpro
import pandas as pd
import plotly.graph_objects as go
from cryptocompy import coin




```

```python
public_client = cbpro.PublicClient()

df=pd.DataFrame(public_client.get_product_historic_rates('ETH-USD', granularity=3600,start='2020-02-17 01:00:00',end='2020-02-29'))
df.columns=['Date','Low','High','Open','Close','Volume']
df['Date']=pd.to_datetime(df['Date'],unit='s')

df.sort_values(by='Date')
```

```python
fig = go.Figure(data=[go.Candlestick(x=df['Date'],
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])

fig.show()
```

```python
finance_charts = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv')


finance_charts.head()
```

```python
coin_data = coin.get_coin_snapshot("ETH","TUSD")
df = pd.DataFrame(coin_data["Exchanges"])
df["VOLUME24HOUR"] = pd.to_numeric(df["VOLUME24HOUR"])
df["VOLUME24HOURTO"] = pd.to_numeric(df["VOLUME24HOURTO"])
df["MEANPRICE"] = df["VOLUME24HOURTO"] / df["VOLUME24HOUR"]
df = df[
    [
        "MARKET",
        "LOW24HOUR",
        "HIGH24HOUR",
        "PRICE",
        "VOLUME24HOUR",
        "VOLUME24HOURTO",
        "MEANPRICE",
    ]
]
df.sort_values(by="VOLUME24HOUR", ascending=False).head()
```
