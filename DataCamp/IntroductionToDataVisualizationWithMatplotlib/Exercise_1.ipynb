```python
fig,ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 3, 3],marker='o',linestyle='--',color='b') 
ax.set_xlabel('X-axis')
ax.set_ylabel('Y1-axis')
ax2 = ax.twinx()
ax2.plot([1, 2, 3, 4], [.8, 7.8, 3.1, 2.8],marker='o',linestyle='--',color='r') 
ax2.set_ylabel('Y2-axis')
```

```python
fig,ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 3, 3],marker='o',linestyle='--',color='b') 
ax.set_xlabel('X-axis')
ax.set_ylabel('Y1-axis')
ax.tick_params('y',colors='blue')
ax2 = ax.twinx()
ax2.plot([1, 2, 3, 4], [.8, 7.8, 3.1, 2.8],marker='o',linestyle='--',color='r') 
ax2.set_ylabel('Y2-axis',color='r')
plt.show()
```

```python
def plot_timeseries(axes,x,y,color,xlabel,ylabel):
    axes.plot(x,y,color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel)
    axes.tick_params('y',colors=color)
    
    
```

```python
fig,ax = plt.subplots()
plot_timeseries(ax,[1, 2, 3, 4], [1, 4, 3, 3],'r','x-axis','y-axis') 
plot_timeseries(ax,[1, 2, 3, 4], [1, 2, 3, 8],'b','x-axis','y-axis') 
```
