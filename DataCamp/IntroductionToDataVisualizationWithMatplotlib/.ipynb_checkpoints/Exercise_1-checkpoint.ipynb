```python
import matplotlib.pyplot as plt
import numpy as np


fig,ax = plt.subplots()
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 3, 3]) 
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 3, 3]) 
ax.plot([1, 2, 3, 4], [.8, 7.8, 3.1, 2.8]) 
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 3, 3],marker='o',linestyle='--',color='b') 
ax.set_xlabel("x axis")
ax.set_ylabel('y axis')
ax.set_title('My first plot')
plt.show()
```
