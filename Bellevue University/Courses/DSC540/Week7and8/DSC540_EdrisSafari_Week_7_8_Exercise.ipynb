```python
MetObjects = pd.read_excel(MetObjects_file)
MetObjects
```

```python
# The unnamed columns are a result of the editing of the files. The Tag column got divided to many columns with no header.
# let's remove them
MetObjects = MetObjects.loc[:, ~MetObjects.columns.str.contains('^Unnamed')]
MetObjects.columns
```

```python
# remove 'Tags' too
MetObjects = MetObjects.drop(columns = ['Tags'])
MetObjects.columns
```

```python
column_names = MetObjects.columns
print('sum\n', MetObjects.isnull()[column_names].sum())
```

```python
remove_columns = MetObjects.columns[MetObjects.isnull().mean() > .8]
print(remove_columns)
```

```python
MetObjects = MetObjects.drop(columns = remove_columns)
MetObjects.columns
```

```python
candy_response_2016 = pd.read_excel(candy_response_2016_file)
```

```python
# These are some arbitrary columns! Some wrangling on the column names will be needed
candy_response_2016.columns
```

```python
# Rename some categorical variables
candy_response_2016 = candy_response_2016.rename(columns={'Your gender:':'Gender','How old are you?':'Age','Are you going actually going trick or treating yourself?':'TrickAndTreating','Which country do you live in?':'Country'})
candy_response_2016.head()
```

```python
# Take a slice of the dataset for evaluation
CandyResponse = pd.DataFrame()
candyResponse = candy_response_2016[['Gender','Age','Country','TrickAndTreating']].dropna()
candyResponse.head()
```

```python
# Multiindex on  Gender and country
candyResponse = candyResponse.set_index(['Gender','Country'])
```

```python
candyResponse.sort_index(level=0).head(20)
```

```python
candyResponse.sort_index(level=1).head(20)
```

```python
# Index heirarchy
candyResponse.index
```

# Chapter 10

```python
candyResponse = candyResponse.reset_index()
candyResponse.head()
```

```python
Gendergroup = candyResponse.groupby(candyResponse['Gender'])
Gendergroup
```

```python
for name, group in Gendergroup:
    print(name)
    print(group)
```

```python
AgeGroup = candyResponse.groupby(['Gender', 'Country'])[['Age']]
```

```python
for name, group in AgeGroup:
    print(group)
    print('\n')
```

```python
CandyResponseWithTimeStamp = pd.DataFrame()
CandyResponseWithTimeStamp = candy_response_2016[['Timestamp','Gender','Age','Country','TrickAndTreating']].dropna()
CandyResponseWithTimeStamp.head()
```

```python
pd.to_datetime(CandyResponseWithTimeStamp.Timestamp)
CandyResponseWithTimeStamp
```

```python
CandyResponseWithTimeStamp["TimeStamp_DMY"] = CandyResponseWithTimeStamp.Timestamp.map(lambda ts: ts.strftime("%d-%m-%Y"))
CandyResponseWithTimeStamp.head()
```

```python
CandyResponseWithTimeStamp = CandyResponseWithTimeStamp.set_index('TimeStamp_DMY')
CandyResponseWithTimeStamp
```

```python
CandyResponseWithTimeStamp.loc['24-10-2016':'25-10-2016:']
```

```python
EarliestResponse = CandyResponseWithTimeStamp.Timestamp.min()
LatestResponse = CandyResponseWithTimeStamp.Timestamp.max()
print(EarliestResponse,LatestResponse) 
```
